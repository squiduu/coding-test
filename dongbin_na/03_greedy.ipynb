```python
# Ch.3 - 그리디 - 01 - 거스름 돈

N = 1260
count = 0
coins = [500, 100, 50, 10]

for coin in coins:
    count += N // coin
    N %= coin

print(count)
```

```python
# Ch.3 - 그리디 - 02 - 1이 될 때까지

N = 25
K = 3
count = 0

while N > 1:
    if N % K == 0:
        N /= K
    else:
        N -= 1
    count += 1

print(count)
```

```python
# Ch.3 - 그리디 - 03 - 곱하기 혹은 더하기

data = input()

result = int(data[0])

for i in range(1, len(data)):
    curr_num = int(data[i])

    if curr_num <= 1 or result <= 1:
        result += curr_num
    else:
        result *= curr_num

print(result)
```

```python
# Ch.3 - 그리디 - 04 - 모험가 길드

N = int(input())
X = list(map(int, input().split()))
X.sort()

num_group = 0
num_people = 0

for i in X:
    num_people += 1
    if num_people >= i:
        num_group += 1
        num_people = 0

print(num_group)
```

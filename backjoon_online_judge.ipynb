```python
# 2,506

n = int(input())
correct = list(map(int, input().split()))
score = [0 for _ in range(len(correct))]

for i in range(len(correct)):
    if i == 0:
        if correct[i] == 0:
            score[i] = 0
        else:
            score[i] = 1
    elif i == 1:
        if correct[i] == 0:
            score[i] = 0
        else:
            score[i] = 1
            if correct[i - 1] == 1:
                score[i] += 1
    elif i >= 2:
        if correct[i] == 0:
            score[i] = 0
        else:
            score[i] = 1
            for j in range(1, i + 1):
                if correct[i - j] == 1:
                    score[i] += 1
                else:
                    break

print(sum(score))
```

```python
# 2,711

t = int(input())

for _ in range(t):
    position, sequence = input().split()
    position = int(position)

    sequence_to_list = []
    new_sequence = ""

    for i in sequence:
        sequence_to_list.append(i)
    for j in range(len(sequence_to_list)):
        if j != (position - 1):
            new_sequence += sequence_to_list[j]
        else:
            new_sequence += ""
    
    print(new_sequence)
```

```python
# 2,822

scores = []
ids = []
for _ in range(8):
    score = int(input())
    scores.append(score)
    sorted_scores = sorted(scores)

for i in range(len(sorted_scores)):
    if scores[i] >= sorted_scores[3]:
        ids.append(i + 1)

print(sum(sorted_scores[3:]))
for j in ids:
    print(j, end=" ")
```

```python
# 2,750

n = int(input())

numbers = []
for _ in range(n):
    numbers.append(int(input()))

for i in range(len(numbers)):
    for j in range(len(numbers)):
        if numbers[i] < numbers[j]:
            numbers[i], numbers[j] = numbers[j], numbers[i]

for n in numbers:
    print(n)
```
